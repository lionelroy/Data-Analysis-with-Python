### Operations in Dataframe

#### Math Operations

```python
import pandas as pd
```

```python
# Reading the csv file
df_volcanoes = pd.read_csv('volcano.csv')
df_volcanoes.head()
```

```python
# Select a column and calculate total sum
print(df_volcanoes['elevation'].sum())
print(df_volcanoes['population_within_5_km'].sum())
print(df_volcanoes['population_within_10_km'].sum())
print(df_volcanoes['population_within_30_km'].sum())
print(df_volcanoes['population_within_100_km'].sum())
```

```python
# count, mean, std, max, and min
print(df_volcanoes['population_within_5_km'].count())
print(df_volcanoes['population_within_5_km'].mean())
print(df_volcanoes['population_within_5_km'].std())
print(df_volcanoes['population_within_5_km'].max())
print(df_volcanoes['population_within_5_km'].min())

print(df_volcanoes['elevation'].count())
print(df_volcanoes['elevation'].mean())
print(df_volcanoes['elevation'].std())
print(df_volcanoes['elevation'].max())
print(df_volcanoes['elevation'].min())
```

```python
# Easier calculation with .describe() method
df_volcanoes.describe()
```

#### Operations in rows

```python
df_annual_earthquake_deaths = pd.read_csv('earthquake_annual_deaths.csv')
df_annual_earthquake_deaths
```

```python

# Calculating the sum in a row
# Earthquake deaths per country from 2000 to 2008
df_deaths_from_1999_to_2008 = df_annual_earthquake_deaths['1999'] + df_annual_earthquake_deaths['2000'] + df_annual_earthquake_deaths['2001'] + df_annual_earthquake_deaths['2002'] + df_annual_earthquake_deaths['2003'] + df_annual_earthquake_deaths['2004'] + df_annual_earthquake_deaths['2005'] + df_annual_earthquake_deaths['2006'] + df_annual_earthquake_deaths['2007'] + df_annual_earthquake_deaths['2008']
df_deaths_from_1999_to_2008
```

```python
# calculating the average earthquake deaths per year whithin a country(2000 - 2008 sum data) and assign the result to a new column
df_annual_earthquake_deaths['10-years average'] = (df_deaths_from_1999_to_2008/10)
```

```python
# showing the dataframe
df_annual_earthquake_deaths.round(2)
```
