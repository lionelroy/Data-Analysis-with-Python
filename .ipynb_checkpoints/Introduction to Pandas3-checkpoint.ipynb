# Introduction to Pandas2

### Value Counts

```python
import pandas as pd
```

```python
# import volcano dataframe
df_volcano = pd.read_csv('volcano.csv')
df_volcano
```

```python
#  Counting country elements
# len function
print(len(df_volcano['country']))
# .count() method
print(df_volcano['country'].count())
```

```python
# Counting elements by category(volcanoes per country)
df_volcano['country'].value_counts()
```

```python
# return the relative frequency (divide all values by the sum of values)
df_volcano['country'].value_counts(normalize=True)
```

```python
# Counting "primary_volcano_type" elements by category
df_volcano['primary_volcano_type'].value_counts()
```

```python
# Return the relative frequency and round to 4 decimals
df_volcano['primary_volcano_type'].value_counts(normalize=True).round(4)
```

### Sort  a dataframe

```python
# Sort by one column
df_volcano.sort_values('last_eruption_year')
```

```python
# Sort descending by one column
df_volcano.sort_values('last_eruption_year', ascending=False)
```

```python
# Sort descending by multiple columns
df_volcano.sort_values(['elevation', 'last_eruption_year'], ascending=False)
```

```python
# Sort descending by multiple columns and update dataframe
# Sort original dataframe by elevation
df_volcano.sort_values(['elevation', 'last_eruption_year'], ascending=False, inplace=True)
```

```python
df_volcano
```

```python
# Sort descending with a key function
# Use Lambda function
df_volcano.sort_values('volcano_name', ascending=True, key=lambda col:col.str.lower())
```
