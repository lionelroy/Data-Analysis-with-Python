### How to create a Dataframe 

#### 1. Create Dataframes with Arrays

```python
# creating an array
data = np.array([[1,4], [2,5], [3,6]])
```

```python
# creating a dataframe(df)
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'],
            columns=['col1', 'col2'])
```

```python
# showing the dataframe
df
```

##### Without using numPy

```python
# creating an array with list shape
data = [[1,4], [2,5], [3,6]]
```

```python
# creating a dataframe(df)
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'],
            columns=['col1', 'col2'])
```

```python
# showing the dataframe
df
```

#### 2. Create a Dataframe from a Dictionary

```python
# Lists used for this example
provinces = ['New Brunswick', 'Quebec', 'Manitoba', 'Nova Scotia', 'Nunavut']
population = [789225, 8604495, 1383765, 992055, 39403]
```

```python
# Storing lists within dictionary
dict_provinces = {'Provinces':provinces, 'Population':population}
```

```python
# Creating the dataframe
df_population = pd.DataFrame(dict_provinces)
```

```python
# Showing the dataframe
df_population
```

### Selecting one column

```python
# Select the column with [] (preferred way to select a column)
df_BAS_naming_conventions['Abbreviation']
```

```python
# Check the data type of a column
type(df_BAS_naming_conventions['Abbreviation'])
```

```python
# series: atributes and methods
df_BAS_naming_conventions['Abbreviation'].index
```

```python
# series: atributes and methods
df_BAS_naming_conventions['Abbreviation'].head()
```

#### Syntax 2(dot notation)

```python
# Select a column with .
df_BAS_naming_conventions.Abbreviation
```

```python
# Select a column with . (pitfalls)
# Can't use variable with space in name
df_BAS_naming_conventions.Haystack Designations
```

```python
# Select the same column using []
df_BAS_naming_conventions['Haystack Designations']
```

### Selecting two or more columns

```python
# Select 2 columns using [[]]
df_BAS_naming_conventions[['Abbreviation', 'Description']]
```

```python
# Check out the data type of the selection
# 1 pair of square brackets will give you a series, 2 pairs will give you a dataframe
type(df_BAS_naming_conventions[['Abbreviation', 'Description']])
```

```python
# selecrt 2 or more columns using [[]]
df_BAS_naming_conventions[['Category', 'Abbreviation', 'Description', 'Sample Use']]
```

```python
# We can't select 2 or more columns with dot notation
# Invalid syntaxt error
df_BAS_naming_conventions.'Abbreviation', 'Description'
```

### Adding a new column with an array

The dataframe has 193 rows so we need to create an array with 193 items

```python
# import numPy
import numpy as np
```

```python
# create an array of 193 elements
# 0 - 193 which is the limit of my range
new_column_data = np.arange(0, 193)
```

```python
# lenght of the array
len(np.arange(0, 193))
```

```python
# Adding a new column to dataframe with an array
df_BAS_naming_conventions['new column'] = new_column_data
df_BAS_naming_conventions
```

```python
#  Create reandom integer numbers between 1 and 100
# random integer method
# first 2 values are range, 3rd argument is the size(number of rows)
int_new_column = np.random.randint(1, 100, size=193)
```

```python
# lenght of the array
len(int_new_column)
```

```python
# min value inclusive and high value exclusive
# second value is exclusive so max will output 99 instead of 100
print(min(int_new_column))
print(max(int_new_column))
```

```python
# adding a new column to dataframe with an array
df_BAS_naming_conventions['new column'] = int_new_column
df_BAS_naming_conventions
```

```python
# create random float numbers between 1 and 100
# use uniform for floats
float_new = np.random.uniform(1, 100, size=193)
```

```python
df_BAS_naming_conventions['new'] = float_new
df_BAS_naming_conventions
```
